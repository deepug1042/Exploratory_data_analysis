# Flight Price Prediction ( EDA + Feature Engineering)

# step1 : import necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# step2: load the train and test dataset

```python
train_data = pd.read_excel("Data_train.xlsx")
train_data.head()
```

```python
test_data = pd.read_excel("Test_set.xlsx")
test_data.head()
```

# step3: combine the 2 dataset

```python
data = pd.concat([train_data,test_data],axis = 0)
```

```python
data.tail()
```

# step4: check for the datatypes

```python
data.info()
```

# step5: select the date_of_journey column and split it using('/') and through indexing

```python
data['Date'] = data['Date_of_Journey'].str.split('/').str[0]
```

```python
data['Month'] = data['Date_of_Journey'].str.split('/').str[1]
data['Year'] = data['Date_of_Journey'].str.split('/').str[2]
```

```python
data.head()
```

# step6: convert the 3 divided column into type int

```python
data['Date'] = data['Date'].astype(int)
data['Month'] = data['Month'].astype(int)
data['Year'] = data['Year'].astype(int)
```

```python
data.info()
```

# step7:drop the date_of_journey column

```python
data.drop('Date_of_Journey',axis = 1,inplace = True)
```

```python
data.head()
```

# step8: now take the arrival_time column

```python
data['Arrival_Time'] = data['Arrival_Time'].str.split(' ').str[0]
```

```python
data.head()
```

# step9: now also we have to split the hour and mins from arrival time

```python
data['Hour'] = data['Arrival_Time'].str.split(':').str[0]
data['Min'] = data['Arrival_Time'].str.split(':').str[1]
```

```python
data.head(2)
```

# step10: drop the arrival time

```python
data.drop("Arrival_Time",axis = 1,inplace = True)
```

```python
data.head()
```

# step10 : split the dep_time and store it sepaerately ,drop the dep_time column

```python
data['Dep_hour'] = data['Dep_Time'].str.split(':').str[0]
data['Dep_min'] = data['Dep_Time'].str.split(':').str[1]
```

```python
data.drop('Dep_Time',axis = 1,inplace = True)
```

```python
data.head(1)
```

```python
data.info()
```

# convert object type to int

```python
data['Hour'] = data['Hour'].astype(int)
data['Min'] = data['Min'].astype(int)
data['Dep_hour'] = data['Dep_hour'].astype(int)
data['Dep_min'] = data['Dep_min'].astype(int)
```

```python
data.info()
```

# step11: convert the total_stops to numbers by mapping function

```python
data['Total_Stops'].unique()
```

```python
#this is the nan row we assign random number to this
data[data['Total_Stops'].isnull()]
```

```python
data['Total_Stops'] = data['Total_Stops'].map({'non-stop': 0,'1 stop':1,'2 stops': 2,'3 stops':3,'nan' : 1})
```

# step12: drop the route column

```python
data.drop('Route',axis = 1,inplace = True)
```

```python
data.head(1)
```

# step13: Now the Duration column should be splitted and converted to mins

```python
data['Duration_hour'] = data['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
data['Duration_min'] = data['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

# now we drop the missplaced value ie 5m from data['Duration_hour']

```python
data[data['Duration_hour']== '5m']
```

```python
#dropping 2 rows which has '5m' as duration
data.drop(6474,axis = 0,inplace = True)
data.drop(2660,axis = 0,inplace = True)
```

```python
data['Duration_hour'] = data['Duration_hour'].astype(int)
```

```python
#converting hours to mins
data['Duration_hour'] = data['Duration_hour']*60
```

```python
#we try to convert the data['Duration_min'] to int
```

```python
data['Duration_min'] = data['Duration_min'].astype(float)
```

```python
data.drop('Duration',axis = 1,inplace = True)
```

```python
data[data['Duration_min'].isnull()]
```

# there is half of the values as nan so we replace it by any thing

```python
data['Duration_min'].mode()[0]
```

```python
data['Duration_min'] = data['Duration_min'].fillna(data['Duration_min'].mode()[0])
```

```python
data['Duration_min'] = data['Duration_min'].astype(int)
```

```python
data.info()
```

# step14: we sum up the converted mins and the mins to one single thing using sum() function"

```python
data['Duration'] = data['Duration_hour'] + data['Duration_min']
```

```python
data.head()
```

```python
#now we can drop the duartion hour and min column
```

```python
data.drop("Duration_min",axis = 1,inplace = True)
data.drop("Duration_hour",axis = 1,inplace = True)
```

```python
data.columns
```

```python
data.head()
```

# step15: now we encode the categorical features using mapping orelese label encoder()

```python
data['Airline'].unique()
```

```python
#we import the lable encoder
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
data['Airline'] = le.fit_transform(data['Airline'])
data['Source'] = le.fit_transform(data['Source'])
data['Destination'] = le.fit_transform(data['Destination'])
data['Additional_Info'] = le.fit_transform(data['Additional_Info'])
```
